```python
import numpy as np                                               #this model is used for the linear matrix
import pandas as pd                                              #used for data processing and for read the file(csv)
import os                                                        #used for upload the file
```

```python
movie_title = pd.read_csv('tmdb_5000_movies.csv')                   #reading the movies file
credit_movies = pd.read_csv('tmdb_5000_credits.csv')                #credit in the movies files
```

```python
movie_title.head(10)                                                 # checking the file which is uploaded
```

```python
movie_title.shape                                               #checking the shape of the file.
```

```python
movies_pro = movie_title.merge(credit_movies,on='title')  # merging the both csv file into 1 using the column name is title
                                                          #(which is similar in both of files)
```

```python
movies_pro.head(10)   # checking up new the file which is created
```

```python
movies_pro = movies_pro[['movie_id','title','overview','genres','keywords','cast','crew']] #taking or creating the file with 
                      #the require data through which the recommend for movies.
```

```python
movies_pro.head(10)                                               # checking up only the data we selected
```

```python
import ast  #used to Analyze the text
```

```python
def convert(text):                     #creating a function to remove the names from the genres
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies_pro.dropna(inplace=True)
```

```python
movies_pro['genres'] = movies_pro['genres'].apply(convert)       #applying the function to the genres and keywords
movies_pro['keywords'] = movies_pro['keywords'].apply(convert)
movies_pro.head(10)                                              #checking up the file
```

```python
def convert3(text):                          #creating a function to remove id , name and space in cast column
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies_pro['cast'] = movies_pro['cast'].apply(convert)                  #appling the function to cast and coverting it into lambda form
movies_pro['cast'] = movies_pro['cast'].apply(lambda x:x[0:3])
movies_pro.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies_pro['crew'] = movies_pro['crew'].apply(fetch_director)
movies_pro.sample(5)
```

```python
def collapse(L):                                     #creating a function to remove space in column
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies_pro['cast'] = movies_pro['cast'].apply(collapse)                  #applying the function and checking it
movies_pro['crew'] = movies_pro['crew'].apply(collapse)
movies_pro['genres'] = movies_pro['genres'].apply(collapse)
movies_pro['keywords'] = movies_pro['keywords'].apply(collapse)
movies_pro.head()
```

```python
movies_pro['overview'] = movies_pro['overview'].apply(lambda x:x.split())
movies_pro['comb'] = movies_pro['overview'] + movies_pro['genres'] + movies_pro['keywords'] + movies_pro['cast'] + movies_pro['crew']
final = movies_pro.drop(columns=['overview','genres','keywords','cast','crew'])
final.head(10)

#applying the lambda to overview
#merging all the columns into one column named as comb
```

```python
final['comb'] = final['comb'].apply(lambda x: " ".join(x))           #covering the comb column into lambda function
final.head()
```

```python
final.to_csv('final_movies.csv',index=False)               #saving the csv into folder
```

```python
from sklearn.feature_extraction.text import CountVectorizer             #import a ml model countvectorizer and giving max words
counter = CountVectorizer(max_features=10000,stop_words='english')  
print(counter)
```

```python
vectorizer = counter.fit_transform(final['comb']).toarray()   #applying the model and toarray
print(vectorizer)                                             #checking up the shape 
vectorizer.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity         #applying the martix and cosine similarity  

similar = cosine_similarity(vectorizer)
similar
```

```python
similar.shape                                                      #checking up the shape
```

```python
similar.size                                                  #checking up the size
```

```python
final[final['title'] == 'The Lego Movie'].index[0]          #see the recommend for the word or seeing the movie location.
```

```python
def recommend(movie):                                            #making function for the recommend for the movies
    movies = final[final['title'] == movie].index[0]
    distance = sorted(list(enumerate(similar[movies])),reverse=True,key = lambda x: x[1])
    for i in distance[1:6]:
        print(final.iloc[i[0]].title)

recommend('Gandhi')                                           #results
```
